```python
b_apt = pd.read_csv('b_apt.csv', encoding='euc-kr')
b_apt= b_apt[['year', 'quarter', 'district', 'b_apt_no_of_cmplx', 'b_apt_avg_space',
       'b_apt_avg_price']]
a=pd.merge(sales, b_apt, on=['year','quarter','district'])
```

```python
s_change = pd.read_csv('s_change.csv', encoding = 'euc-kr')
b = pd.merge(a,s_change,on=['year', 'quarter', 'district'])
```

```python
b_facil = pd.read_csv('b_facil.csv', encoding = 'euc-kr')
b_facil=b_facil[['year', 'quarter', 'district', 'b_facil_total','b_facil_no_of_STA', 'b_facil_no_of_BUS']]
c = pd.merge(b,b_facil,on=['year', 'quarter', 'district'])
```

```python
s_float_pop = pd.read_csv('s_float_pop.csv', encoding = 'euc-kr')
d = pd.merge(c,s_float_pop,on=['year', 'quarter', 'district'])
```

```python
b_resid = pd.read_csv('b_resid.csv', encoding = 'euc-kr')
e = pd.merge(d,b_resid,on=['year', 'quarter', 'district'])
```

```python
s_store = pd.read_csv('s_store.csv', encoding = 'euc-kr')
s_store = s_store[['year', 'quarter', 'district', 'code', 's_store_no_of_store']]
f = pd.merge(e,s_store, on= ['year', 'quarter', 'district', 'code'])
```

```python
s_worker = pd.read_csv('s_worker.csv', encoding = 'euc-kr')
g = pd.merge(f, s_worker, on= ['year', 'quarter', 'district'])
```

```python
b_income = pd.read_csv('b_income.csv', encoding = 'euc-kr')
h = pd.merge(g, b_income, on=['year', 'quarter', 'district'])
```

```python
h=h[['year', 'quarter', 'district', 'code',  'sales_total','sales_weekday_amt',
       'sales_female_amt', 'sales_2030s_amt', 'sales_06_11_amt',
       'sales_11_14_amt', 'sales_14_17_amt', 'sales_17_21_amt',
       'sales_21_24_amt', 'b_apt_no_of_cmplx',
       'b_apt_avg_space', 'b_apt_avg_price', 'open_rate', 'close_rate',
       'b_facil_total', 'b_facil_no_of_STA', 'b_facil_no_of_BUS', 'floatpop',
       'b_resid_total', 'b_total_fam', 'b_resid_no_of_apt_fam',
       'b_resid_no_of_nonapt_fam', 's_store_no_of_store', 's_worker',
       'income_avg', 'expand_avg']]
```

```python
df_merged=h.groupby(['year','quarter','district','code','sales_total']).agg({
    'sales_weekday_amt': 'sum',
    'sales_female_amt': 'sum',
    'sales_2030s_amt': 'sum',
    'sales_06_11_amt': 'sum',
    'sales_11_14_amt': 'sum',
    'sales_14_17_amt': 'sum',
    'sales_17_21_amt': 'sum',
    'sales_21_24_amt': 'sum',
    'b_apt_no_of_cmplx': 'mean',
    'b_apt_avg_space': 'mean',
    'b_apt_avg_price': 'mean',
    'open_rate': 'mean',
    'close_rate': 'mean',
    'b_facil_total': 'mean',
    'b_facil_no_of_STA': 'mean',
    'b_facil_no_of_BUS': 'mean',
    'floatpop': 'mean',
    'b_resid_total': 'mean',
    'b_total_fam': 'mean',
    'b_resid_total': 'mean',
    'b_resid_no_of_apt_fam': 'mean',
    'b_resid_no_of_nonapt_fam': 'mean',
    's_store_no_of_store': 'mean',
    's_worker': 'mean',
    'income_avg': 'mean',
    'expand_avg': 'mean'
})
```

### s_float_pop 변수 만들기

```python
s_float_pop = pd.read_csv('s_floatPop.csv', encoding = 'euc-kr')
```

```python
s_float_male_col=s_float_pop.columns[3:255]
```

```python
s_float_female_col= s_float_pop.columns[256:]
```

```python
s_float_20s = []
s_float_30s = []
s_float_40s = []

for i in s_float_pop.columns:
    if "20s" in i:
        s_float_20s.append(i)
        
    elif "30s" in i:
        s_float_30s.append(i)
        
    elif "40s" in i:
        s_float_40s.append(i)
        
```

```python
s_float_pop[s_float_20s].tail()
```

```python
s_float_pop['s_float_male'] = s_float_pop[s_float_pop.columns[3:255]].T.sum()
```

```python
s_float_pop['s_float_female'] = s_float_pop[s_float_female_col].T.sum()
```

```python
s_float_pop['s_float_20s'] = s_float_pop[s_float_20s].T.sum()
s_float_pop['s_float_30s'] = s_float_pop[s_float_30s].T.sum()
s_float_pop['s_float_40s'] = s_float_pop[s_float_40s].T.sum()
```

```python
s_float_pops=pd.concat([s_float_pop[s_float_pop.columns[:3]],s_float_pop[s_float_pop.columns[-5:]]], axis=1)
```
