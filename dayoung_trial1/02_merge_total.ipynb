```python
#각 파일 불러와서 primary 키 중복되는 데이터 있는지 확인 후 추정매출데이터와 merge
```

```python
class merge_sales:
    
    def __init__(self,file):
        self.file = file
    
    def read(self):
        self.name = pd.read_csv("{}".format(self.file),encoding='utf-8')
        return self.name.tail(3)
    
    def show_columns(self):
        return self.name.columns
    
    def change_columns(self,new_columns):
        
        self.name.columns = new_columns
        return self.name.tail(3)
    
    def cut(self,g_columns):
        self.name = self.name[g_columns]
        return self.name.tail(3)
    
    def check(self,keys):
        check = self.name[self.name.duplicated(keys)]
        return check
    
    def save(self):
        self.name.to_csv("{}".format(self.file), sep=",", index=False)

        

    
```
