```python
#상권 데이터 
s_df = pd.read_csv("sang_income.csv", encoding='euc-kr')
s_df.head()
```

```python
s_df_nan=s_df.copy()
```

```python
#0인 경우 NAN으로 바꾸기
def make_nan(x):
    if x==0:
        return np.nan
    else :
        return x
    
for col in s_df_nan.columns :
    s_df_nan[col] = s_df_nan[col].apply(make_nan)

s_df_nan[30:40]
```

```python
import missingno as msno
msno.matrix(s_df_nan)
plt.show()
```

```python
#nan값이 하나라도 있는 행은 없애기
s_df_nan=s_df_nan.dropna(axis=0)
msno.bar(s_df_nan)
plt.show()
```

```python
s_df_nan["employee_rate"]=s_df_nan['여성_직장_인구_수']/(s_df_nan['남성_직장_인구_수']+s_df_nan['여성_직장_인구_수'])
s_df_nan["sales"]=s_df_nan['당월_매출_금액']/s_df_nan['점포_수']
s_df_nan["floating_pop_rate"]= s_df_nan['여성_유동인구_수']/(s_df_nan['남성_유동인구_수']+s_df_nan['여성_유동인구_수'])
```
