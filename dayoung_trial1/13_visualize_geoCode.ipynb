## grs80tm좌표계를 WGS84 좌표계로 변환하기

```python
district_code = pd.read_csv('district_code.csv', encoding='utf-8' )
district_code=district_code[:100]
# district_code['상권_코드'].min(), district_code['상권_코드'].max()
# for i in district_code['상권_코드'].unique(): 

district_code.상권_코드
```

```python
import requests

# WGS84 or WCONGNAMUL or CONGNAMUL or WTM or TM
lat_lng = []
for x,y,code in zip (district_code['엑스좌표_값'],district_code['와이좌표_값'],district_code['상권_코드'] ):
    x_ori = x
    y_ori = y
    url = "https://dapi.kakao.com/v2/local/geo/transcoord.json?x={}&y={}&input_coord=TM&output_coord=WGS84".format(x_ori,y_ori)

    headers = {
        "Host": "dapi.kakao.com",
        "Authorization": "KakaoAK e85e7f7718701e158bc96ae1a736a64e",

    }

    response= requests.get(url, headers=headers)
    data = response.json()["documents"]
    for i in data:
        x_trans, y_trans, code= i['x'],i['y'] , code
        lat_lng.append([y_trans,x_trans, code] )
lat_lng
```
