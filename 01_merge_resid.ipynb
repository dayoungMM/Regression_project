```python
#매출 데이터 
sales = pd.read_csv("sales_2018.csv", encoding='euc-kr')
sales.head(10)
```

```python
monthly_sale = sales_merge[sales_merge.columns[:5]]
monthly_sale.to_csv('monthly_sale.csv',sep=",",index= False, encoding='euc-kr')
```

#### 당월매출금액/점포수 컬럼 가지는 df 만들기
- 만약 다월 매출금액이 점포 1개당 매출금액이 아닐경우 사용하기 위함

```python
sales_merge=sales_merge[['기준_년_코드','기준_분기_코드','상권_코드','서비스_업종_코드','당월_매출_금액','점포수']]
sales_merge.tail(3)
```

```python
sales_merge['sales']=round(sales_merge['당월_매출_금액']/sales_merge['점포수'] , 0)
sales_merge.tail()
```

```python
sale1 = sales_merge[sales_merge.columns[:-3]]
```

```python
sale2 = sales_merge['sales']
```

```python
sales_result = pd.concat([sale1,sale2], axis=1)
sales_result.tail()
```

```python
#저장
sales_result.to_csv('sales_result.csv',sep=",",index= False, encoding='euc-kr')
```

```python
#동일 기간, 동일 상권인데 매출이 다른 행이 존재한다는것을 확인
monthly_sale[monthly_sale.duplicated(['기준_년_코드','기준_분기_코드','상권_코드'])]
```

### 상주인구

```python
resid = pd.read_csv("resid.csv", encoding='euc-kr')
resid.head(1)
```

```python
resid_merge = resid.copy()
resid_merge.columns
```

```python
resid_merge=resid_merge[['기준_년_코드', '기준_분기_코드','상권 코드','총 상주인구 수', '총 가구 수']]
resid_merge.tail()
```

```python
resid_merge = resid_merge[resid_merge['기준_년_코드']==2018]
resid_merge.tail()
```

```python
pd.Series.value_counts(resid_merge['상권 코드'])
```

```python
#동일 기간, 동일 상권인데 인구가 다른 중복행이 존재하지 않는다
resid_merge[resid_merge.duplicated(['기준_분기_코드','상권 코드'])]
```

```python
resid_merge.to_csv("resid.csv",  sep=",", encoding = 'euc-kr', index=False)
```

### 매출과 상주인구 데이터 합치기

```python
#merge하기 위해 key값 이름 동일하기 맞추기위해 column명 변경
resid_merge.columns = ['기준_년_코드', '기준_분기_코드', '상권_코드', '총 상주인구 수', '총 가구 수']
```

```python
df=pd.merge(monthly_sale, resid_merge, on=['기준_년_코드', '기준_분기_코드', '상권_코드'])
```
